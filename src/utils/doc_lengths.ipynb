### Msmarco Docs

```python
columns = ['doc_id', 'number_of_words']
```

```python
doc_lengths_docs = pd.read_csv("/gpfs/home3/sgarcarz/PycharmProjects/ColBERT/src/data/msmarco_docs_doc_lengths.tsv", sep='\t', header=None, names=columns)
```

```python
doc_lengths_docs['number_of_words'].plot.hist(grid=True, bins=10, rwidth=0.9, color='#607c8e')
```

```python
data = doc_lengths_docs['number_of_words']

# Specify bin edges
bin_edges = list(np.linspace(0, 512, 11))  # 10 bins between 0 and 512
bin_edges.append(float('inf'))  # add an upper bound for the "512 and more" bin

# Plotting the histogram
plt.hist(data, bins=bin_edges, rwidth=0.9, color='#607c8e', align='left')
plt.grid(True)

# Labeling and showing the plot
plt.xlabel('Number of Words')
plt.ylabel('Count')
plt.title('Histogram of Word Counts')
plt.xticks(bin_edges[:-1], rotation=45)  # Setting the x-ticks to the bin starts for clarity
plt.tight_layout()
plt.show()
```

```python
# Your sample data (assuming it's a pandas Series)
data = doc_lengths_docs['number_of_words']

bin_edges = list(np.linspace(0, 512, 11))  # 10 bins between 0 and 512
bin_edges.append(float('inf'))  # >512

# Get histogram counts using specified bin edges
counts, _ = np.histogram(data, bins=bin_edges)


# Create a DataFrame to hold the table values
table = pd.DataFrame({
    'Range': [f"{bin_edges[i]:.2f} - {bin_edges[i+1]:.2f}" for i in range(len(bin_edges)-1)],
    '# Docs': counts
})
table
```
